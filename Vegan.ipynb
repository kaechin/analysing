```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('vegandata.csv', encoding='ISO-8859-1')
df.columns

#Repalce to question number when questions words are longer than 30
newlist = df.columns.to_list()
for i in range(len(newlist)):
    if len(newlist[i]) > 30:
        newlist[i] = 'q%d' % i
        
df.columns = np.array(newlist)

#Line graph for Age and Gender
sns.kdeplot(data=df, x="Age", hue= "Gender")
plt.show()

#Pie chart for Gender
df[['Gender']].groupby("Gender").size().plot.pie()
plt.show()

#Bar graph for Gender and Level of Education
sns.countplot('Gender',hue='Level of Education',data=df)
plt.show()
```
